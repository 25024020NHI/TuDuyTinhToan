### lam tron mot so thanh co k chu so sau dau phap


```python
a= float(input())
k= int(input())
b= a*(10**k)
if (b*10-int(b)*10)>=5 :
    c= int(b + 1) / (10**k)
    print(c)
else :
    d= int(b) / (10**k)
    print(d)
```
