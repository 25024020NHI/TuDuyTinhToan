#### Nhập vào số nguyên dương n, tính tổng các chữ số của n

#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n=int(input())
a=0
b=0
while n>0:
    if (n%10)%2==0:
        a+=1
    else:
        b+=1
    n=n//10
print('co', a, 'chu so chan', b, 'chu so le')
```

#### Nhập n Cho S(k) = 1 + 2 + 3 + … + k Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n=int(input())
while True:
    if n<1:
        n=int(input())
        continue
    else:
        break
k =int((-1+(1+8*n)**0.5)/2)
if k*(k+1)==2*n:
    k=k-1
print(k)
```

#### Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A= int(input())
while True:
    if A<1:
        A=int(input())
        continue
    else:
        break
s=0
n=0
while s<=A:
    n+=1
    s+=1/n
print(n)
```

#### Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
b=0
c=0
while True:
    a=int(input())
    if a==-1:
        break
    if a>b:
            b=a
    if a<c:
            c=a
print('max =', b, 'min =', c)
```

#### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
a = int(input())
f1, f2 = 1, 1
while f2 <= a:
    f1, f2 = f2, f1+f2
print(f1)
```

#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
chuoi=input()
a=0
for i in chuoi:
    if i!=" ":
        a+=1
print(a)
```

#### Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
chuoi=input()
for index, i in enumerate(chuoi):
    if i ==' ':
        print(chuoi[:index])
        break
```

#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số 

```python
chuoi=input()
a=0
b=0
c=0
for i in chuoi:
    if i.isupper():
        a+=1
    elif i.islower():
        b+=1
    elif i.isdigit():
        c+=1
print('co', a, 'in hoa', b, 'in thuong', c, 'so')
```
