### Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def amstr(x):
    str_x = str(x)
    n=len(str_x)
    tong=0
    for i in str_x:
        tong+=int(i)**n
    return tong==x
x=int(input())
if amstr(x):
    print(x, "la so amstronhg")
else:
    print("false")
```

### VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def sodoixung(x):
    a=0
    b=x
    while x>0:
        a=a*10+x%10
        x=x//10
    return b==a
x=int(input())
if sodoixung(x):
    print("TRUE")
else:
    print("false")

```

### VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def sokydieu(x):
    while x>=10:
        tong=0
        while x>0:
            tong+=x%10
            x=x//10
        x=tong
    return x==1
x=int(input())
if sokydieu(x):
    print("true")
else:
    print("false")

```

### Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.


```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        pass
    try:
        float(s)
        return 'float'
    except ValueError:
        return 'neither'
s=input()
print(detect_number_type(s))
```

###  Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại


```python
def is_one_product_of_others(a,b,c):
    return a==b*c or b==a*c or c==a*b
a,b,c=map(int,input().split())
if is_one_product_of_others(a,b,c):
    print("True")
else:
    print("False")
```

### Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

### Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    return (x2-x1)/(y2-y1)==(x3-x1)/(y3-y1)
x1,y1,x2,y2,x3,y3=map(float,input().split())
if are_collinear(x1,y1,x2,y2,x3,y3):
    print('yes')
else:
    print('no')
```
