##### Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n,m = map(int,input().split())
matr = []
for i in range(n):
    row = list(map(int,input().split()))
    matr.append(row)
for row in matr:
    for val in row:
        print(f'{val:4}', end='')
    print()
    
```

##### Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list

```python
lst=list(map(int,input().split()))
def maxxx(lst):
    a=lst[0]
    for i in lst:
        if i>a:
            a=i
    return a
print (maxxx(lst))
```

##### Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
lst=list(map(int,input().split()))
k=int(input())
def tim_vi_tri(lst,k):
    for i in range (len(lst)):
        if lst[i]==k:
            return i
    if k not in lst:
        return -1
print(tim_vi_tri(lst,k))
        
```

##### Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

##### VIết hàm xóa đi các số <x có trong mảng

```python
lst= list(map(int,input().split()))
x=int(input())
lon_hon_x=[]
def ham_lon_hon_x(lst,x):
    for i in lst:
        if i<=x:
            lon_hon_x.append(i)
    return lon_hon_x
print(ham_lon_hon_x(lst,x))
```

##### Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
lst= list(map(int,input().split()))
bang_khong=[]
khac_khong=[]
def dau_khong(lst):
    for i in lst:
        if i == 0:
            bang_khong.append(i)
        if i != 0:
            khac_khong.append(i)
    return bang_khong+khac_khong
print(dau_khong(lst))
                                  
```

##### Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
lst=list(map(int,input().split()))
def tong_cuc_dai_dia_phuong(lst):
    a=0
    for i in range (1,len(lst)-1):
        if lst[i-1]<lst[i]>lst[i+1]:
            a+=lst[i]
    return a
print(tong_cuc_dai_dia_phuong(lst))
```

##### Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

```python
lst=list(map(int,input().split()))
x=int(input())
def nho_lon_x(lst,x):
    a=[]
    b=0
    for i in lst:
        if i<x:
            a.append(i)
            b+=1
    if b==0:
        return '-1'
    return a
print(nho_lon_x(lst,x))
```

##### Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
lst=list(map(int,input().split()))
def kiem_tra(lst):
    tang= True
    giam=True
    for i in range (len(lst)-1):
        if lst[i]>=lst[i+1]:
            tang=False
        if lst[i]<=lst[i+1]:
            giam=False
    if tang:
        return 'tang'
    if giam:
        return 'giam'
    return 'vo danh'
print(kiem_tra(lst))
        
        
```
